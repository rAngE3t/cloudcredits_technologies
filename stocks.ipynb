```python

import pandas as pd
import yfinance as yf
from datetime import datetime

# Tech stocks
tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN']

# Time range
end = datetime.now()
start = datetime(end.year - 1, end.month, end.day)

company_list = []
company_name = ["APPLE", "GOOGLE", "MICROSOFT", "AMAZON"]

for stock, name in zip(tech_list, company_name):
    df = yf.download(stock, start=start, end=end)
    df["company_name"] = name
    company_list.append(df)

# Combine all data
final_df = pd.concat(company_list, axis=0)

final_df.tail(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.1)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Close'].plot()   # use 'Close' instead of 'Adj Close'
    plt.ylabel('Close Price')
    plt.xlabel("")
    plt.title(f"Closing Price of {tech_list[i - 1]}")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.1)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Volume'].plot()
    plt.ylabel('Volume')
    plt.xlabel("")
    plt.title(f"Sales Volume for {tech_list[i - 1]}")

plt.tight_layout()
plt.show()
```

```python
ma_day = [10, 20, 50]

# Calculate moving averages using 'Close'
for ma in ma_day:
    for company in company_list:
        company[f"MA_{ma}"] = company['Close'].rolling(window=ma).mean()

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for ax, company, name in zip(axes.flatten(), company_list, tech_list):
    company[['Close', 'MA_10', 'MA_20', 'MA_50']].plot(ax=ax)
    ax.set_title(name)

plt.tight_layout()
plt.show()

```

```python
# Calculate daily return
for company in company_list:
    company['Daily Return'] = company['Close'].pct_change()

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for ax, company, name in zip(axes.flatten(), company_list, tech_list):
    company['Daily Return'].plot(
        ax=ax,
        legend=True,
        linestyle='--',
        marker='o'
    )
    ax.set_title(name)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 9))

for i, (company, name) in enumerate(zip(company_list, company_name), 1):
    plt.subplot(2, 2, i)
    company['Daily Return'].dropna().hist(bins=50)
    plt.xlabel('Daily Return')
    plt.ylabel('Counts')
    plt.title(name)

plt.tight_layout()
plt.show()
```

```python
data = df['Close']

dataset = data.values
training_data_len = int(np.ceil( len(dataset) * .95 ))
training_data_len
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
scaled_data
```

```python
train_data = scaled_data[0:int(training_data_len), :]
x_train = []
y_train = []
for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print() 
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

model = Sequential()

# Smaller LSTM layers
model.add(LSTM(50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(25, return_sequences=False))

model.add(Dense(10))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

model.fit(
    x_train,
    y_train,
    batch_size=32,   
    epochs=5,
    verbose=1
)
```

```python
test_data = scaled_data[training_data_len - 60:, :]

x_test = []
y_test = dataset[training_data_len:, :]

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
predictions = model.predict(x_test)
```
